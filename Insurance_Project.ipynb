```python
test_ids = test_data['id']
test_data['premium'] = None #bc thats what we're predicting
#merge them
merged_data = pd.concat([train_data, test_data], ignore_index=True)
```

```python
missing = merged_data.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)

print("Missing values per column:\n", missing)
```

```python
numeric_cols = merged_data.select_dtypes(include='number').columns
for col in numeric_cols:
    if merged_data[col].isnull().sum() > 0:
        # fill the numeric values with mean
        merged_data[col] = merged_data[col].fillna(merged_data[col].mean())
```

```python
obj_cols = merged_data.select_dtypes(include='object').columns
for col in obj_cols:
    if merged_data[col].isnull().sum() > 0:
        # get the most frequent item
        modes = merged_data[col].mode()
        # fill the missing value with the first most frequent value -- modes[0]
        if not modes.empty:
            mode_val = modes[0]
            merged_data[col] = merged_data[col].fillna(mode_val)
        else:
            # if the entire column is missing, fill in unknown
            merged_data[col] = merged_data[col].fillna("Unknown")
```

```python
#check
print("Remaining missing values:", merged_data.isnull().sum().sum())
```
